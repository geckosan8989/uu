#### Cleaning Title Text by using Regex library

```python
def clean_text(text):
    #removal of url
    text = re.sub(r'https?://\S+|www\.\S+|http?://\S+',' ',text) 
    
    #removal of html tags
    text = re.sub(r'<.*?>',' ',text) 
    
    text = re.sub("["
                           u"\U0001F600-\U0001F64F"  # removal of emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+",' ',text)
    
    
    #remove newline char
    text = re.sub('\n', '', text)
    
    return text
```

```python
books['Book-Title'] = books['Book-Title'].apply(lambda x:clean_text(x))
```

## Collaborative Filtering

```python
x= ratings_with_name.groupby("User-ID").count()["Book-Rating"] >250
x[x].index
```

```python
exp_users=x[x].index #we checked
```

```python
filtered_rating=ratings_with_name[ratings_with_name["User-ID"].isin(exp_users)] #ther did 10 lakhs rating out of which 4.74 lakhs did the rating by exp_users 
```

```python
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
```

```python
famous_books=y[y].index
```

```python
final1_ratings=filtered_rating[filtered_rating["Book-Title"].isin(famous_books)] #books who are only in famous books
```

```python
final1_ratings.drop_duplicates()
```

```python
pt=final1_ratings.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index=np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6] #to find maximum similarity score in descending order
   
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:11]
    
    data=[]
    for i in similar:
        item=[]
        temp=final_book_data[final_book_data["Book-Title"]==pt.index[i[0]]]
        item.extend(list(temp.drop_duplicates("Book-Title")["Image-URL-M"]))
        item.extend(list(temp.drop_duplicates("Book-Title")["Book-Title"]))
        item.extend(list(temp.drop_duplicates("Book-Title")["Book-Author"]))
        data.append(item)
    return data
```
